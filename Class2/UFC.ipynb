```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('UFC large_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Handling null values

```python
df.isnull().sum()
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
# Identify object columns
object_cols = df.select_dtypes(include=['object']).columns

# Replace null values in object columns with their mode
df.fillna({col: df[col].mode()[0] for col in object_cols}, inplace=True)
```

```python
# Identify float columns
float_cols = df.select_dtypes(include=['float']).columns

# Replace null values in float columns with their median
df.fillna({col: df[col].median() for col in float_cols}, inplace=True)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df.isnull().sum().sum()
```

```python
for col in df.columns:
    unique_values = df[col].unique()
    if len(unique_values) < 10:  # Condition for less than 10 unique values
        print(f"Column: {col}")
        
        print(unique_values)
        print("-" * 40)
```

## Visualization

#### Bar Chart

```python
plt.figure(figsize = (8,6))
pclass_count = df["method"].value_counts()
pclass_count.plot(kind = 'bar', color = 'red', edgecolor = 'black') 
plt.title('Count of KO Method')
plt.xlabel('KO Method')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

#### Histogram

```python
plt.figure(figsize = (8,6))
plt.hist(df['sub_avg_diff'], bins = 28, color = 'green', edgecolor = 'black') #bins -> interval gap
plt.title('sub_avg_diff distribution')
plt.xlabel('sub_avg_diff')
plt.ylabel('frequency')
plt.show()
```

#### Bar Chart

```python
plt.figure(figsize = (8,6))
s_count = df["winner"].value_counts()
name = ['won', 'lose']
plt.pie(s_count, labels = name, autopct = '%1.3f%%', startangle = 90, colors = ['skyblue','red'])
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("winner distribution")
plt.show()
```

#### Scatter Plot

```python
plt.figure(figsize = (8,6))
plt.scatter(df['td_avg_diff'], df['time_sec'], c = 'skyblue', edgecolor = 'k', alpha = 0.5)
# alpha -> to control the transparency
# k -> black color
plt.title('td_avg_diff vs Time')
plt.xlabel('td_avg_diff')
plt.ylabel('Time (s)')
plt.show()
```

#### Line Chart

```python
plt.figure(figsize = (8,6))
plt.plot(df.index, df['r_str'], c = 'orange', lw = 2)
# lw -> line width
plt.title('sub_avg_diff distribution')
plt.xlabel('Index')
plt.ylabel('sub_avg_diff')
plt.show()
```

#### BoxPlot

```python
plt.figure(figsize = (8,6))
df.boxplot(column = 'time_sec', by = 'winner', grid = False, patch_artist = True)
# patch_artist -> to fill the boxplot with color
plt.suptitle('')
plt.title('Boxplot distribution')
plt.xlabel('winner')
plt.ylabel('Time (s)')
plt.show()
# Higer viscus, lower viscus, median(greenline), outliers
```

#### Donut Chart

```python
df['is_title_bout'].value_counts()
```

```python
plt.figure(figsize = (8,6))
s_count = df["is_title_bout"].value_counts()
name = ['no', 'yes']
plt.pie(s_count, labels = name, autopct = '%1.2f%%', startangle = 90, colors = ['skyblue','red'], wedgeprops=dict(width=0.4))
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("Is match a title bout ?")
plt.show()
```
