```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Handling Null values

```python
df.isnull().sum()
```

```python
null_per = df.isnull().mean() * 100
print(null_per)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.head()
```

## Visualization

```python
a = df["Pclass"].value_counts() # value of unique values
print(a)
```

#### Bar Graph (category / frequency / comparison)

```python
plt.figure(figsize = (8,6))
pclass_count = df["Pclass"].value_counts()
pclass_count.plot(kind = 'bar', color = 'skyblue', edgecolor = 'black') 
plt.xlabel('Passenger class')
plt.ylabel('Count')
plt.xticks(rotation = 0)
plt.show()
```

#### Histogram(continous value  / distribution )

```python
plt.figure(figsize = (8,6))
plt.hist(df['Age'], bins = 28, color = 'green', edgecolor = 'black') #bins -> interval gap
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

#### Pie Chart(category / segments)

```python
plt.figure(figsize = (8,6))
s_count = df["Survived"].value_counts()
name = ['Not Survived', 'Survived']
plt.pie(s_count, labels = name, autopct = '%1.3f%%', startangle = 90, colors = ['skyblue', 'pink'])
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("Survival distribution")
plt.show()
```

#### Scatter Plot(relation between two numerical col)

```python
plt.figure(figsize = (8,6))
plt.scatter(df['Age'], df['Fare'], c = 'skyblue', edgecolor = 'k', alpha = 0.5)
# alpha -> to control the transparency
# k -> black color
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

#### Line Chart(trends)

```python
plt.figure(figsize = (8,6))
plt.plot(df.index, df['Fare'], c = 'orange', lw = 2)
# lw -> line width
plt.title('Fare distribution')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.show()
```

#### BoxPlot (distribution / two columns / first colm numeric / second col category)

```python
plt.figure(figsize = (8,6))
df.boxplot(column = 'Fare', by = 'Pclass', grid = False, patch_artist = True)
# patch_artist -> to fill the boxplot with color
plt.suptitle('')
plt.title('Boxplot distribution')
plt.xlabel('P class')
plt.ylabel('Fare')
plt.show()
# Higer viscus, lower viscus, median(greenline), outliers
```

#### Donut Chart(same as  pie chart )

```python
plt.figure(figsize = (8,6))
s_count = df["Survived"].value_counts()
name = ['Not Survived', 'Survived']
plt.pie(s_count, labels = name, autopct = '%1.3f%%', startangle = 90, colors = ['skyblue', 'pink'], wedgeprops=dict(width=0.4))
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("Survival distribution")
plt.show()
```
